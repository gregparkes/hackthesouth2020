```python
#Finding all labelnames for a given image ID
tmp = df3.groupby('ImageID')['LabelName'].apply(lambda x: x.str.cat(sep=';'))
```

```python
df1.set_index("ImageID", inplace=True)
```

```python
#Adding label name column to image data frame
df1['LabelName'] = tmp
```
